```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pp
```

```python
u_d = pd.read_csv("/content/user_demographics_v1.csv")
```

```python
u_d[(u_d.device=='and')&(u_d.age>15)&(u_d.age<30)]
```

```python
u_d[(u_d.device=='iOS')&(u_d.age>15)&(u_d.age<30)]
```

```python
c1 = len ( u_d[(u_d.device=='and')&(u_d.age>15)&(u_d.age<30)] )
c2 = len ( u_d[(u_d.device=='iOS')&(u_d.age>15)&(u_d.age<30)] )
```

```python
pp.title ("Pop. b/w 15 - 30")
pp.bar("and",c1)
pp.bar("iOS",c2)
pp.plot()
pp.show()
```
