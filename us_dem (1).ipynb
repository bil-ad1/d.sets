```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pp
```

```python
u_d = pd.read_csv("/content/user_demographics_v1.csv")
```

```python
a1 = len( u_d[(u_d.country=='USA')&(u_d.gender=='M')&(u_d.device=='and')] )
a3 = len ( u_d[(u_d.country=='USA')&(u_d.gender=='M')&(u_d.device=='iOS')] )
```

```python
pp.bar("US-M-and",a1)
pp.bar("US-M-iOS",a3)
pp.plot()
pp.show()
```

```python
a2 = len ( u_d[(u_d.country=='USA')&(u_d.gender=='F')&(u_d.device=='and')] )
a4 = len ( u_d[(u_d.country=='USA')&(u_d.gender=='F')&(u_d.device=='iOS')] )
```

```python
pp.bar("US-F-and",a2)
pp.bar("US-F-iOS",a4)
pp.plot()
pp.show()
```
